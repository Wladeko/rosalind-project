```python
import re
```

```python
with open('rosalind_prot.txt') as f:
    s = f.read().replace('\n', '')
```

```python
with open('codon.txt') as f:
    codon = f.read().replace('\n', ' ')

codon = re.sub(' +',' ',codon).split(' ')[:-1]
for i in range(len(codon)):
    codon[i] = codon[i].replace('Stop', ' ')

codon = {codon[2*i]: codon[2*i+1] for i in range(len(codon)//2)}
```

```python
n = 3
s_chunks = [s[i:i+n] for i in range(0, len(s), n)]

protein = [codon[i] for i in s_chunks]
protein = ''.join(protein)
protein
```
